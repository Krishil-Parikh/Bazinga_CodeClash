

 <h2 style="text-align:center;color:#D35400"> Model Building</h2>

```python
X = tr.drop(columns=['Target'])
Y = tr['Target']
```

```python
# Encode "Target" Varibale using label encoder

le = LabelEncoder()
tr['Target'] = le.fit_transform(tr['Target'])
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# initialize the model for classification
rd = RandomForestClassifier(n_estimators=100, random_state=42)

# fit the model
rd.fit(X_train, y_train)

# make predictions
y_pred = rd.predict(X_test)

# evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
# import classification report

from sklearn.metrics import classification_report
```

```python


# Initialize the models
cb = CatBoostClassifier(n_estimators=100, random_state=42, silent=True)
lgbm = LGBMClassifier(n_estimators=100, random_state=42)
SvM = SVC(kernel='linear', decision_function_shape='ovr', probability=True)

# Fit the models
cb.fit(X_train, y_train)
lgbm.fit(X_train, y_train)
SvM.fit(X_train, y_train)


# Make predictions
y_pred_cb = cb.predict(X_test)
y_pred_lgbm = lgbm.predict(X_test)
y_pred_svm=SvM.predict(X_test)

# Evaluate the models
accuracy_cb = accuracy_score(y_test, y_pred_cb)
accuracy_lgbm = accuracy_score(y_test, y_pred_lgbm)
accuracy_svm=accuracy_score(y_test,y_pred_svm)

print("Accuracy of CatBoost: {:.2f}%".format(accuracy_cb * 100))
print("Accuracy of LightGBM: {:.2f}%".format(accuracy_lgbm * 100))
print("Accuracy of SVM: {:.2f}%".format(accuracy_svm * 100))

```

```python
report=classification_report(y_test,y_pred_svm)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred_svm)
print("Confusion Matrix:")
print(conf_matrix)
```
